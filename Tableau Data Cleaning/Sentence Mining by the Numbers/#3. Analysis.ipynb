# Libraries

```python
#required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
colors = ['red', 'blue', 'purple', 'orangered', 'salmon', 'magenta', 'green']

vocab_sizes = [2000]

for vocab_size_init in vocab_sizes:
    
    print("Initial Vocab Size:", vocab_size_init)

    #load the relevant data for the relevant vocab size
    learned_words_data_HP = pd.read_csv('Harry Potter/learned_words_data_' + str(vocab_size_init)  +  '.csv')
    known_words_HP = pd.read_csv('Harry Potter/known_words_' + str(vocab_size_init)  +  '.csv', index_col=0)
    learned_words_HP = known_words_HP.loc[vocab_size_init+1:]
    learned_words_HP = learned_words_HP.loc[learned_words_HP['known'] ==1,]
    
    #just for fun
    learned_words_data_HP['Log_Med_Freq'] = np.log(learned_words_data_HP['Med_Freq'])
    
    
    #---VOCAB GROWTH---
    fig, ax1 = plt.subplots(1, 1, sharey=True)

    fig.set_figheight(10)
    fig.set_figwidth(15)


    for book_num in range(7):

        ax1.plot(learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1,].index, learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1, 'Vocab_After'], color=colors[book_num])

    fig.suptitle('Cumulative Vocabulary Growth', fontsize=25)

    plt.show()
    
    
    #--WORDS LEARNED---
    fig, ax1 = plt.subplots(1, 1, sharey=True)

    fig.set_figheight(7)
    fig.set_figwidth(20)

    for book_num in range(7):

        ax1.plot(learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1,].index, learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1, 'Learned_Words'], color=colors[book_num])

    fig.suptitle('Words learned each Chapter', fontsize=25)

    plt.show()
    
    
    #---DIFFICULTY---
    fig, ax1 = plt.subplots(1, 1, sharey=True)

    fig.set_figheight(7)
    fig.set_figwidth(20)

    for book_num in range(7):

        ax1.plot(learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1,].index, learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1, 'Perc_Unknown'], color=colors[book_num])

    fig.suptitle('Difficulty of each Chapter', fontsize=25)

    plt.show()
    
    
    #---EFFICIENCY---
    fig, ax1 = plt.subplots(1, 1, sharey=True)

    fig.set_figheight(7)
    fig.set_figwidth(20)

    for book_num in range(7):

        ax1.plot(learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1,].index, learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1, 'Learned_per_Unknown'], color=colors[book_num])

    fig.suptitle('Efficiency of each Chapter', fontsize=25)

    plt.show()
    
    
    #---Rareness---
    fig, ax1 = plt.subplots(1, 1, sharey=True)

    fig.set_figheight(7)
    fig.set_figwidth(20)

    for book_num in range(7):

        ax1.plot(learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1,].index, learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1, 'Med_Freq'], color=colors[book_num])

    fig.suptitle('Median Learned-Word Frequency per Chapter', fontsize=25)

    plt.show()
    
    
    #---Log Rareness---
    fig, ax1 = plt.subplots(1, 1, sharey=True)

    fig.set_figheight(7)
    fig.set_figwidth(20)

    for book_num in range(7):

        ax1.plot(learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1,].index, learned_words_data_HP.loc[learned_words_data_HP['Book']==book_num+1, 'Log_Med_Freq'], color=colors[book_num])

    fig.suptitle('Log Median Learned-Word Frequency per Chapter', fontsize=25)

    plt.show()
    
    
    print('\n\n\n\n\n')
```
